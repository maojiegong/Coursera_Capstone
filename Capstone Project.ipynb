Exercise1: Scrape Wikipedia Page and Create Dataframe
Will scrape the table from Wikipedia website first and then processing the data as required:
1. There are 3 tables totally and the first table is what we need
2. Check and remove boroughs that "Not assigned"
3. Check if there is neighbourhood "Not assigned" and result is that that is no "Not assigned" neighbourhood
4. Check if there are duplicate postal codes and the result is no duplicate postal code exist
5. After remove "Not assigned" boroughs the index need to be reset
6. Show total postal codes in Torondo using .shape method

```python
#Import libraries
import pandas as pd
import numpy as np
#print("Hello Capstone Project Course!")
```

```python
#Scrape the Wikipedia page to get the table.
table_Toronto =pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')

#Check the number of tables
len(table_Toronto)

#Look through all the tables
table_Toronto

#Get the needed table
df = table_Toronto[0]

#Remove Borough that "Not assigned"
df_new = df.loc[df.Borough != "Not assigned"]

#Check Neighbourhood "Not assigned"
#df_new_neig = df_new.loc[df_new.Neighbourhood =="Not assigned"]
#df_new_neig
#There is no neighbourhood that are "Not Assigned", so the df_new will be final output

#Reset index and make default index as index and remove original index
df_new.reset_index(inplace = True, drop = True)
df_new

#Check if there are duplicated postal codes in different rows
#df_new[df_new.duplicated(['Postal Code'])]
#No duplicated postal code found

#Show total number of postal codes
df_new.shape
```

Exercise2: Get the latitude and the longitude coordinates of each neighborhood. 
I met some issue using geocoder and therefore just used CSV file for coordinate info.
To simply merge two files together to get the consolidated file with coordinate info added. 

```python
import pandas as pd
import numpy as np

#Read csv file with coordinate info
df_csv = pd.read_csv('https://cocl.us/Geospatial_data')
df_csv

#consolidate two files together to have coordinate info added to each neighbour. 
df_com = pd.merge(df_new, df_csv, how = 'outer', on = ['Postal Code'])
df_com
```

```python
#Use geopy library to get the latitude and longitude values of Downtown Toronto
address = 'Downtown Toronto'

geolocator = Nominatim(user_agent="tr_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Downtown Toronto are {}, {}.'.format(latitude, longitude))
```

```python
#Create a map of Toronto with neighborhoods superimposed on top.
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_com['Latitude'], df_com['Longitude'], df_com['Borough'], df_com['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
